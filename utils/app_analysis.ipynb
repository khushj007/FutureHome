```python
f_df = pd.read_csv('/content/gurgaon_properties_cleaned_v1.csv')
```

```python
f_df.head()
```

```python
f_df["sector"].unique()
```

```python
f_df.replace({"sohna road road":"sohna road","sector 3 phase 3 extension":"sector 3","sector 3 phase 2 extension":"sector 3","sector 3 phase 2":"sector 3"},inplace=True)
```

```python
f_df = f_df[["sector","features"]]
```

```python
f_df
```

```python
f_df.dropna(inplace=True)
```

```python
f_df["features"] = f_df["features"].apply(lambda x: eval(x))
```

```python
f_df
```

```python
def feature_group_togather(sector):
  value = []
  feature_list = f_df[f_df["sector"] == sector]["features"].values
  for val in feature_list:
    value = value+val
  return value
```

```python
features = pd.DataFrame()
```

```python
for sector in f_df["sector"].unique():
  value = {
      "sector":sector,
      "features":[feature_group_togather(sector)]
  }
  features = pd.concat([features,pd.DataFrame(value,index=[0])])

```

```python
import joblib

joblib.dump(features,"features.pkl")
```
